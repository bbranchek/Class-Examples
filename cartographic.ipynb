```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Natural Earth countries (110m scale) hosted on GitHub
url = "https://raw.githubusercontent.com/nvkelso/natural-earth-vector/master/geojson/ne_110m_admin_0_countries.geojson"

world = gpd.read_file(url)

# Compute population density if fields exist
if "POP_EST" in world.columns and "AREA" in world.columns:
    world["pop_density"] = world["POP_EST"] / world["AREA"]
else:
    # fallback if fields differ
    world["pop_density"] = 0

fig, ax = plt.subplots(figsize=(12, 6))
world.plot(
    column="pop_density",
    cmap="viridis",
    legend=True,
    ax=ax,
    edgecolor="black",
    linewidth=0.3
)

ax.set_title("World Population Density (Natural Earth GeoJSON)")
plt.show()
```

```python
import pandas as pd
from shapely.geometry import Point

cities = pd.DataFrame({
    "city": ["New York", "Los Angeles", "Chicago"],
    "lon": [-74.0060, -118.2437, -87.6298],
    "lat": [40.7128, 34.0522, 41.8781]
})

gdf = gpd.GeoDataFrame(
    cities,
    geometry=gpd.points_from_xy(cities.lon, cities.lat),
    crs="EPSG:4326"
)

fig, ax = plt.subplots(figsize=(10, 6))
world.plot(ax=ax, color="lightgray", edgecolor="black")
gdf.plot(ax=ax, color="red", markersize=80)

for _, row in gdf.iterrows():
    ax.text(row.geometry.x + 1, row.geometry.y + 1, row.city)

ax.set_title("Major US Cities")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# --- Synthetic climate field (example: global temperature anomaly) ---
# In real NASA workflows, this would come from:
# MERRA-2, GEOS-5, GISS, ERA5, CMIP6, etc.

lons = np.linspace(-180, 180, 360)
lats = np.linspace(-90, 90, 181)
lon2d, lat2d = np.meshgrid(lons, lats)

# Example climate anomaly pattern (synthetic)
data = 2.5 * np.sin(3 * np.deg2rad(lat2d)) * np.cos(2 * np.deg2rad(lon2d))

# --- NASA-style projection ---
proj = ccrs.Orthographic(central_longitude=-30, central_latitude=20)

fig = plt.figure(figsize=(12, 12))
ax = plt.axes(projection=proj)

# --- Satellite background (Blue Marble) ---
ax.stock_img()

# --- Plot climate field ---
# Use PlateCarree for input data
pcm = ax.pcolormesh(
    lons,
    lats,
    data,
    transform=ccrs.PlateCarree(),
    cmap="coolwarm",          # NASA-style anomaly colormap
    shading="auto",
    alpha=0.65                # blend with satellite imagery
)

# --- Scientific overlays ---
ax.add_feature(cfeature.COASTLINE, linewidth=0.7, edgecolor="white")
ax.add_feature(cfeature.BORDERS, linewidth=0.5, edgecolor="white")

# --- Gridlines ---
gl = ax.gridlines(
    draw_labels=False,
    linewidth=0.4,
    color="white",
    alpha=0.4,
    linestyle="--"
)

# --- Colorbar (NASA style) ---
cbar = plt.colorbar(
    pcm,
    orientation="horizontal",
    pad=0.05,
    shrink=0.8
)
cbar.set_label("Temperature Anomaly (°C)", fontsize=12)

# --- Title ---
ax.set_title(
    "Global Climate Field (NASA-Style Template)\nSynthetic Temperature Anomaly",
    fontsize=18,
    pad=20
)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# -----------------------------
# 1. Create global lat/lon grid
# -----------------------------
lons = np.linspace(-180, 180, 360)
lats = np.linspace(-90, 90, 181)
lon2d, lat2d = np.meshgrid(lons, lats)

# -----------------------------
# 2. Synthetic climate fields
# -----------------------------

# A) Global temperature anomaly
temp_anom = 2.5 * np.sin(3 * np.deg2rad(lat2d)) * np.cos(2 * np.deg2rad(lon2d))

# B) Aerosol Optical Depth (regional)
aod = np.exp(-((lat2d - 35)**2 + (lon2d + 95)**2) / 800)

# C) Sea ice concentration (Arctic)
ice = np.clip(1 - (lat2d - 90)**2 / 2500, 0, 1)

# D) Sea Surface Temperature
sst = 15 + 10 * np.cos(np.deg2rad(lat2d))

# -----------------------------
# 3. Create multi-panel figure
# -----------------------------
fig = plt.figure(figsize=(16, 16))

# Panel A — Orthographic (Global Temperature Anomaly)
ax1 = plt.subplot(2, 2, 1, projection=ccrs.Orthographic(-30, 20))
ax1.stock_img()
pcm1 = ax1.pcolormesh(lons, lats, temp_anom,
                      transform=ccrs.PlateCarree(),
                      cmap="coolwarm", shading="auto", alpha=0.65)
ax1.add_feature(cfeature.COASTLINE, edgecolor="white")
ax1.add_feature(cfeature.BORDERS, edgecolor="white")
ax1.set_title("A) Global Temperature Anomaly", fontsize=14)

# Panel B — Lambert Conformal (AOD)
ax2 = plt.subplot(2, 2, 2, projection=ccrs.LambertConformal(-95, 35))
ax2.set_extent([-125, -66.5, 24, 50], crs=ccrs.PlateCarree())
ax2.stock_img()
pcm2 = ax2.pcolormesh(lons, lats, aod,
                      transform=ccrs.PlateCarree(),
                      cmap="inferno", shading="auto", alpha=0.6)
ax2.add_feature(cfeature.COASTLINE, edgecolor="white")
ax2.add_feature(cfeature.BORDERS, edgecolor="white")
ax2.add_feature(cfeature.STATES, edgecolor="white")
ax2.set_title("B) Aerosol Optical Depth (AOD)", fontsize=14)

# Panel C — Polar Stereographic (Sea Ice)
ax3 = plt.subplot(2, 2, 3, projection=ccrs.NorthPolarStereo())
ax3.set_extent([-180, 180, 50, 90], crs=ccrs.PlateCarree())
ax3.stock_img()
pcm3 = ax3.pcolormesh(lons, lats, ice,
                      transform=ccrs.PlateCarree(),
                      cmap="Blues", shading="auto", alpha=0.7)
ax3.add_feature(cfeature.COASTLINE, edgecolor="white")
ax3.set_title("C) Arctic Sea Ice Concentration", fontsize=14)

# Panel D — Robinson (SST)
ax4 = plt.subplot(2, 2, 4, projection=ccrs.Robinson())
ax4.stock_img()
pcm4 = ax4.pcolormesh(lons, lats, sst,
                      transform=ccrs.PlateCarree(),
                      cmap="coolwarm", shading="auto", alpha=0.6)
ax4.add_feature(cfeature.COASTLINE, edgecolor="white")
ax4.add_feature(cfeature.BORDERS, edgecolor="white")
ax4.set_title("D) Global Sea Surface Temperature", fontsize=14)

# -----------------------------
# 4. Add shared colorbars
# -----------------------------
cbar_ax1 = fig.add_axes([0.25, 0.48, 0.5, 0.015])
cbar1 = fig.colorbar(pcm1, cax=cbar_ax1, orientation="horizontal")
cbar1.set_label("Temperature Anomaly (°C)")

cbar_ax2 = fig.add_axes([0.25, 0.03, 0.5, 0.015])
cbar2 = fig.colorbar(pcm4, cax=cbar_ax2, orientation="horizontal")
cbar2.set_label("Sea Surface Temperature (°C)")

plt.tight_layout(rect=[0, 0.05, 1, 1])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# ============================================================
# NASA TOOLKIT MODULE
# ============================================================

def nasa_colorbar(fig, mappable, label, position="bottom"):
    if position == "bottom":
        cax = fig.add_axes([0.25, 0.05, 0.5, 0.015])
    else:
        cax = fig.add_axes([0.25, 0.92, 0.5, 0.015])

    cbar = fig.colorbar(mappable, cax=cax, orientation="horizontal")
    cbar.set_label(label, fontsize=12)
    cbar.ax.tick_params(labelsize=10)
    return cbar


def nasa_panel_label(ax, label):
    ax.text(
        0.02, 0.98, label,
        transform=ax.transAxes,
        fontsize=16,
        fontweight="bold",
        va="top",
        ha="left",
        color="white"
    )


def nasa_annotation(ax, text, loc="lower left"):
    if loc == "lower left":
        x, y = 0.02, 0.02
    elif loc == "upper left":
        x, y = 0.02, 0.98
    elif loc == "upper right":
        x, y = 0.98, 0.98
    else:
        x, y = 0.98, 0.02

    ax.text(
        x, y, text,
        transform=ax.transAxes,
        fontsize=11,
        color="white",
        ha="left" if "left" in loc else "right",
        va="bottom" if "lower" in loc else "top",
        bbox=dict(facecolor="black", alpha=0.3, pad=4)
    )


def nasa_scale_bar(ax, length_km, location=(0.1, 0.05)):
    x, y = location
    ax.text(
        x, y + 0.02,
        f"{length_km} km",
        transform=ax.transAxes,
        color="white",
        fontsize=10,
        ha="left"
    )
    ax.plot(
        [x, x + 0.15], [y, y],
        transform=ax.transAxes,
        color="white",
        linewidth=3
    )


def nasa_inset_globe(fig, parent_ax, lon=0, lat=0, size=0.18):
    inset_ax = fig.add_axes([
        parent_ax.get_position().x1 - size - 0.01,
        parent_ax.get_position().y1 - size - 0.01,
        size, size
    ], projection=ccrs.Orthographic(lon, lat))

    inset_ax.stock_img()
    inset_ax.add_feature(cfeature.COASTLINE, linewidth=0.4, edgecolor="white")
    inset_ax.set_global()
    return inset_ax


# ============================================================
# SYNTHETIC CLIMATE DATA (runs anywhere)
# ============================================================

lons = np.linspace(-180, 180, 360)
lats = np.linspace(-90, 90, 181)
lon2d, lat2d = np.meshgrid(lons, lats)

temp_anom = 2.5 * np.sin(3 * np.deg2rad(lat2d)) * np.cos(2 * np.deg2rad(lon2d))
aod = np.exp(-((lat2d - 35)**2 + (lon2d + 95)**2) / 800)
ice = np.clip(1 - (lat2d - 90)**2 / 2500, 0, 1)
sst = 15 + 10 * np.cos(np.deg2rad(lat2d))


# ============================================================
# MULTI-PANEL NASA FIGURE
# ============================================================

fig = plt.figure(figsize=(16, 16))

# Panel A — Global Temperature Anomaly
ax1 = plt.subplot(2, 2, 1, projection=ccrs.Orthographic(-30, 20))
ax1.stock_img()
pcm1 = ax1.pcolormesh(lons, lats, temp_anom,
                      transform=ccrs.PlateCarree(),
                      cmap="coolwarm", shading="auto", alpha=0.65)
ax1.add_feature(cfeature.COASTLINE, edgecolor="white")
ax1.add_feature(cfeature.BORDERS, edgecolor="white")
nasa_panel_label(ax1, "A")
nasa_annotation(ax1, "Global Temperature Anomaly")
nasa_inset_globe(fig, ax1, lon=-30, lat=20)

# Panel B — AOD over North America
ax2 = plt.subplot(2, 2, 2, projection=ccrs.LambertConformal(-95, 35))
ax2.set_extent([-125, -66.5, 24, 50], crs=ccrs.PlateCarree())
ax2.stock_img()
pcm2 = ax2.pcolormesh(lons, lats, aod,
                      transform=ccrs.PlateCarree(),
                      cmap="inferno", shading="auto", alpha=0.6)
ax2.add_feature(cfeature.COASTLINE, edgecolor="white")
ax2.add_feature(cfeature.BORDERS, edgecolor="white")
ax2.add_feature(cfeature.STATES, edgecolor="white")
nasa_panel_label(ax2, "B")
nasa_annotation(ax2, "Aerosol Optical Depth (AOD)", loc="upper left")
nasa_scale_bar(ax2, 1000)

# Panel C — Arctic Sea Ice
ax3 = plt.subplot(2, 2, 3, projection=ccrs.NorthPolarStereo())
ax3.set_extent([-180, 180, 50, 90], crs=ccrs.PlateCarree())
ax3.stock_img()
pcm3 = ax3.pcolormesh(lons, lats, ice,
                      transform=ccrs.PlateCarree(),
                      cmap="Blues", shading="auto", alpha=0.7)
ax3.add_feature(cfeature.COASTLINE, edgecolor="white")
nasa_panel_label(ax3, "C")
nasa_annotation(ax3, "Arctic Sea Ice Concentration")

# Panel D — Global SST
ax4 = plt.subplot(2, 2, 4, projection=ccrs.Robinson())
ax4.stock_img()
pcm4 = ax4.pcolormesh(lons, lats, sst,
                      transform=ccrs.PlateCarree(),
                      cmap="coolwarm", shading="auto", alpha=0.6)
ax4.add_feature(cfeature.COASTLINE, edgecolor="white")
ax4.add_feature(cfeature.BORDERS, edgecolor="white")
nasa_panel_label(ax4, "D")
```
